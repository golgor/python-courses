# Filtering dataframes

```python
# Filtering is about rows
# Creating a filter
filter1 = stats.InternetUsers < 2 # Creates a list of True / False values according to the evaluation done
filter2 = stats.BirthRate > 40

# Or do it in one row
stats[stats.InternetUsers < 2]
```

```python
# Combining filters does not work with keywoard "and". "and" assumes a single value, not a list or array of True / False
stats[filter1 and filter2]
```

```python
# Need to use '&' as it is bitwise
stats[filter1 & filter2]
```

```python
# Gives error due to order of operations!
stats[stats.InternetUsers < 2 & stats.BirthRate > 40]
```

```python
# Fix with parenteses
stats[(stats.InternetUsers < 2) & (stats.BirthRate > 40)]
```

```python
stats[stats.IncomeGroup == "High income"]
```

```python
# How to get all different values in a column. Unique lists all different values
stats.IncomeGroup.unique()
```

```python
stats[stats.CountryName == "Malta"]
```

```python
stats[stats.CountryName == "Sweden"]
```

# Using .at() and .iat()

```python
# Accessing individual elements
# .at # for labels, even integers are treated as labels
# .iat # for integer location
```

```python
# Accessing row 3 and column 4
stats.iat[3,4]
```

```python
# using labels instead of numbers for columns
stats.at[0, "BirthRate"]
```

```python
sub10 = stats[::10]
print(sub10)
```

```python
# .iat counts up to row number 10
sub10.iat[10,0]
```

```python
# While using at treats the number as a label, i.e. index = 10
sub10.at[10,"CountryName"]
```
