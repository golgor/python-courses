## Subsetting columns

```python
stats.head()
```

```python
# Using strings, it will automatically infer columns as columns have "names"
stats["CountryCode"]
```

```python
stats["CountryName"].head()
```

```python
# Pass in a list of column names to show several columns. That is why there is 2 brackets
stats[["CountryName", "BirthRate"]].head()
```

```python
# Object notation works as well. Important to keep column names without spaces if you want to use this
# Works to get access to one column only
stats.CountryName
```

## Subsetting with combinations

```python
# Like regular slicing with 2d-arrays as well... array[][]
stats[4:8][["CountryName", "BirthRate"]]
```

```python
# Above is creating two subsets at the same time, i.e.
df1 = stats[4:8]
df1 = df1[["CountryName", "BirthRate"]]
print(df1)
```

```python
# Order does not matter, compare with above
print(stats[["CountryName", "BirthRate"]][4:8])
```
