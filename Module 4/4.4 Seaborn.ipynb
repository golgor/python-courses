```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
stats = pd.read_csv("P4-Demographic-Data.csv")
stats.columns = ["CountryName", "CountryCode", "BirthRate", "InternetUsers", "IncomeGroup"]
```

```python
plt.rcParams["figure.figsize"] = 8,4
vis1 = sns.distplot(stats["InternetUsers"])
sns.set(style="darkgrid")
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
```

```python
vis2 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

```python
# As seaborn are used to combine several plots in the same window, the rcParam does not work to increase size.
# These plots have their own setting, size and aspect
vis3 = sns.lmplot(data = stats, x = "InternetUsers", y = "BirthRate",
                  fit_reg = False, hue = "IncomeGroup", size = 7, aspect = 2)
```

# Keyword arguments

```python
#Example using marker size for the graph above
# Seaborn are relying on matplotlib. 
#"scatter,line}_kws : dictionaries - Additional keyword arguments to pass to ``plt.scatter`` and ``plt.plot``.
# Seaborn is a wrapper for matplotlib.pyplot.scatter
vis3 = sns.lmplot(data = stats, x = "InternetUsers", y = "BirthRate",
                  fit_reg = False, hue = "IncomeGroup", size = 7, aspect = 2, scatter_kws = {"s":80})
```
