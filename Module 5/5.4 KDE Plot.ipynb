```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
movies = pd.read_csv("P4-Movie-Ratings.csv")
movies.columns = ["Film", "Genre", "CriticRating", "AudienceRating", "Budget", "Year"]
```

```python
sns.set(style="darkgrid")
vis1 = sns.lmplot(data = movies, x = "CriticRating", y = "AudienceRating", fit_reg = False, hue = "Genre", size = 7, aspect = 1.5)
```

```python
# Kernel Density Estimate
k1 = sns.kdeplot(movies.CriticRating, movies.AudienceRating)
```

```python
k1 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade = True, shade_lowest = False, cmap = "Reds")
```
