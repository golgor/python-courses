```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
movies = pd.read_csv("P4-Section6-Homework-Dataset.csv")
sns.set(style="darkgrid")
```

```python

# Change column names
movies.columns =["Weekday", "Director", "Genre", "MovieTitle", "ReleaseDate", "Studio", "AdjustedGross", "Budget", "Gross", "IMDBRating", "MovielensRating", "Overseas", "Overseas_p", "Profit", "Profit_p", "Playtime", "US", "GrossUS_p"]
```

```python
movies.info()
```

```python
# Change category types
movies.Weekday = movies.Weekday.astype("category")
movies.Director = movies.Director.astype("category")
movies.Genre = movies.Genre.astype("category")
movies.Studio = movies.Studio.astype("category")
```

```python
movies.info()
```

```python
movies.describe()
```

```python
# Lists with what genres and studios that is of interest
genres = ["action", "comedy", "adventure", "animation", "drama"]
studios = ["Buena Vista Studios", "Sony", "Universal", "WB", "Paramount Pictures", "Fox"]

# Filters for genres and studios
movies = movies[movies.Genre.isin(genres)]
movies = movies[movies.Studio.isin(studios)]

# Remove unused categories
movies.Genre.cat.remove_unused_categories(inplace = True)
movies.Studio.cat.remove_unused_categories(inplace = True)

# Print
print(movies.Genre.cat.categories)
print(movies.Studio.cat.categories)
```

```python
# Create boxplot
fig, ax = plt.subplots()
fig.set_size_inches(14.7, 8.27)

# Box plot
box1 = sns.boxplot(data = movies, x = "Genre", y = "GrossUS_p", palette = "inferno")

# Scatter plot
kwargs = {"s":150}
scatter1 = sns.scatterplot(data = movies, x = "Genre", y = "GrossUS_p", hue = "Studio", **kwargs)

plt.legend(loc = "upper left", bbox_to_anchor = (1, 1))
```
