```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
movies = pd.read_csv("P4-Movie-Ratings.csv")
movies.columns = ["Film", "Genre", "CriticRating", "AudienceRating", "Budget", "Year"]
```

```python
sns.set(style="dark")
k1 = sns.kdeplot(movies.Budget, movies.AudienceRating)
```

```python
k2 = sns.kdeplot(movies.Budget, movies.CriticRating)
```

```python
f, axes = plt.subplots(1,2, figsize = (12,6))
k1 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0])
k2 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[1])
```

```python
# If 2 dimensional subplots, axes becomes a 2D-array, i.e. we have to specify axes as "axes[][]"
f, axes = plt.subplots(2,2, figsize = (12,6))
k1 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0][1])
k2 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[1][0])
```

```python
# Change limits by set xlim. Set for one and let all plots share the same limits by "sharex = True" and/or "sharey = True"
f, axes = plt.subplots(1,2, figsize = (12,6), sharex = True, sharey = True)
k1 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0])
k2 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[1])
k1.set(xlim = (-20, 180))
```

```python
# If 2 dimensional subplots, axes becomes a 2D-array, i.e. we have to specify axes as "axes[][]"
f, axes = plt.subplots(2,2, figsize = (12,6),sharex = True, sharey = True)
k1 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0][1])
k2 = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[1][0])
```
