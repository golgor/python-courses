```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
movies = pd.read_csv("P4-Movie-Ratings.csv")
movies.columns = ["Film", "Genre", "CriticRating", "AudienceRating", "Budget", "Year"]
```

```python
# Style is set on the sns object, i.e. all plots using sns
sns.set(style="darkgrid")
m1 = sns.distplot(movies.AudienceRating, bins = 15)
```

```python
m2 = sns.distplot(movies.CriticRating, bins = 15)
```

Notice the difference in distributions between audience and critics. Audience more normally distributed, while critics are more uniform.

```python
n1 = plt.hist(movies.AudienceRating, bins = 16)
```

Looks different from seaborn

```python
# When plotting using matplotlib the columns tend to hide behind each other
plt.hist(movies[movies.Genre == "Action"].Budget, bins = 15)
plt.hist(movies[movies.Genre == "Drama"].Budget, bins = 15)
plt.hist(movies[movies.Genre == "Thriller"].Budget, bins = 15)
```

```python
budget = [movies[movies.Genre == "Action"].Budget, movies[movies.Genre == "Drama"].Budget, movies[movies.Genre == "Thriller"].Budget]
plt.hist(budget, bins = 15, stacked = True)
plt.show()
```

```python
# Change certain columns type to category
movies.Film = movies.Film.astype("category")
movies.Genre = movies.Genre.astype("category")
movies.Year = movies.Year.astype("category")
movies.info()
```

```python
movie_plot = list()
for genre in movies.Genre.cat.categories:
    movie_plot.append(movies[movies.Genre == genre].Budget)
plt.hist(movie_plot, bins = 30, stacked = True, rwidth = 0.9, label = movies.Genre.cat.categories)
plt.legend()
plt.show()
```
