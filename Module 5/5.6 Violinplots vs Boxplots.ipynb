```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
movies = pd.read_csv("P4-Movie-Ratings.csv")
movies.columns = ["Film", "Genre", "CriticRating", "AudienceRating", "Budget", "Year"]
```

```python
vio1 = sns.violinplot(data = movies, x = "Genre", y = "CriticRating")
```

```python
box1 = sns.boxplot(data = movies, x = "Genre", y = "CriticRating")
```

```python
box1 = sns.boxplot(data = movies[movies.Genre == "Drama"], x = "Year", y = "CriticRating")
```

```python
box1 = sns.violinplot(data = movies[movies.Genre == "Drama"], x = "Year", y = "CriticRating")
```

See difference between the boxplot and violinplot for 2008 and 2011. In the boxplot they look pretty much similar, but in violinplot it is clear that the distributions are different.
